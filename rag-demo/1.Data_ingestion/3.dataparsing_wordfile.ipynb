### Word Document Processing

```python
from langchain_community.document_loaders import Docx2txtLoader,UnstructuredWordDocumentLoader
```

```python
## mETHOD1: Using Docx2txtLoader
print("METHOD1: Using Docx2txtLoader")

try:
    docx_loader = Docx2txtLoader("data/word_files/sample.docx")
    docs = docx_loader.load()
    print(f"Number of documents loaded: {len(docs)}")
    print(f"Content of the first document:\n{docs[0].page_content[:500]}")  # Print first 500 characters
    print(f"Metadata of the first document:\n{docs[0].metadata}")

    for i, doc in enumerate(docs):
        print(f"\n--- Document {i+1} ---")
        print(f"Content:\n{doc.page_content[:500]}")  # Print first 500 characters
        print(f"Metadata:\n{doc.metadata}")

except Exception as e:
    print(f"Error loading .docx file with Docx2txtLoader: {e}")
```

```python
print("METHOD2: Using UnstructuredWordDocumentLoader")

## METHOD2: Using UnstructuredWordDocumentLoader
try:
    unstructured_loader = UnstructuredWordDocumentLoader("data/word_files/sample.doc")
    docs_unstructured = unstructured_loader.load()
    print(f"Number of documents loaded: {len(docs_unstructured)}")
    print(f"Content of the first document:\n{docs_unstructured[0].page_content[:500]}")  # Print first 500 characters
    print(f"Metadata of the first document:\n{docs_unstructured[0].metadata}")
except Exception as e:
    print(f"Error loading .docx file with UnstructuredWordDocumentLoader: {e}")
```
