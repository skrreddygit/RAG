#### Create our First Enmbeddings using HuggingFace

```python
#Example code to generate sentence embeddings
from sentence_transformers import SentenceTransformer

sentences = ["This is an example sentence","Each Sentence is converted"]

model=SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
embeddings=model.encode(sentences)
#print(embeddings) 
```

```python
### HuggingFace and Open AI Models

from langchain_huggingface import HuggingFaceEmbeddings


## Initialize HuggingFace Embeddings (No API Key Required)
## First Train Your model
huggingface_embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
print(huggingface_embeddings)

```

```python
# Create Your First Embeddings Here

text = "This is a sample text to be embedded."
vector_embeddings = huggingface_embeddings.embed_query(text)
print(f"Embedding Length: {len(vector_embeddings)}")
print(vector_embeddings)

```

```python
sentences=["The Cat Sat on the mat",
           "The dog sat on the log",
           "Cats and dogs are great pets",
           "I love to play football",
           "Football is a great sport"
           "Python is a programming language",
           "I Love coding in Python"]


vector_sentence_embeddings = huggingface_embeddings.embed_documents(sentences)
print(f"Number of Sentence Embeddings: {len(vector_sentence_embeddings)}")
print(vector_sentence_embeddings[0])

print("--------")
print("\nDetailed Sentence Embeddings:")
for i, sentence_embedding in enumerate(vector_sentence_embeddings):
    print(f"Sentence {i+1} embedding length: {len(sentence_embedding)}")
    print(sentence_embedding)
```
